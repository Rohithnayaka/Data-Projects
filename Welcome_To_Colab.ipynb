# Crime Data Analysis Report

This report summarizes the key findings from the Exploratory Data Analysis (EDA) performed on the `Crime_Data.csv` dataset. The analysis aimed to understand the dataset's characteristics, identify prevalent crime types, analyze the distribution of 'Commits' (likely representing offenses or individuals involved), and investigate temporal trends in crime data.

## 1. Data Overview and Cleaning

*   **Dataset Size:** The dataset comprises 24,596 crime records with 6 columns: `Complaint Number`, `Major Crime Head`, `Crime Head and Section`, `Minor Crime Head`, `Commits`, and `Month`.
*   **Data Types:** `Complaint Number` is an integer, `Commits` is a float, and other columns are object (string) types.
*   **Missing Values:**
    *   Missing values in `Crime Head and Section` were identified and filled with 'Unknown'.
    *   Missing values in `Commits` were imputed using the median value (0.0), due to the skewed nature of its distribution.
*   **Date Conversion:** The `Month` column was successfully converted to datetime objects to facilitate time-series analysis.

## 2. Distribution of Crime Types

### Major Crime Head
*   **Most Prevalent:** 'KARNATAKA STATE LOCAL ACTS' is the most frequent major crime head, indicating a high number of incidents related to local regulations or laws.
*   **Other Significant Categories:** 'THEFT', 'KIDNAPPING AND ABDUCTION', and 'Murder' also represent significant portions of the major crime incidents.

### Minor Crime Head
*   **Dominant Category:** 'Others' is the most frequent minor crime head, suggesting that a large number of crimes fall into less specific categories or are not detailed further in the dataset.
*   **Next Most Frequent:** 'Attempt to Commit' is the next most common specific minor crime head.

## 3. Distribution of 'Commits'

*   **Skewed Distribution:** The 'Commits' column exhibits a highly right-skewed distribution, with a strong concentration of values at the lower end (around 0).
*   **Outliers:** The presence of numerous outliers with exceptionally high 'Commits' values was observed. These outliers might correspond to major crime incidents or events involving multiple offenses or individuals. This skewness suggests that most crime incidents involve a small number of 'Commits', while a few incidents are associated with a very large number.

## 4. Relationship Between Crime Type and 'Commits'

*   **High Average Commits:** 'CHEATING', 'MOTOR VEHICLE ACCIDENTS NON-FATAL', and 'CYBER CRIME' were identified as the 'Major Crime Heads' with the highest average number of 'Commits' (all averaging over 400). This indicates that these crime types tend to involve more associated 'Commits', possibly due to their nature (e.g., complex schemes in cheating, multiple parties in accidents, or widespread impact in cybercrime).

## 5. Temporal Trends

*   **Overall Monthly Trend:** The overall monthly trend of 'Commits' shows fluctuations over time, with a notable dip around April 2020, followed by recovery and varying patterns in subsequent periods.
*   **Trends by Major Crime Head:** Breaking down the monthly trends by top 5 'Major Crime Heads' revealed diverse temporal patterns. Some crime categories might contribute more significantly to overall spikes or dips than others.
*   **Trends by Minor Crime Head:** Further granular analysis by top 5 'Minor Crime Heads' provided detailed insights into the specific monthly fluctuations of these crime types, which can be valuable for targeted interventions.

## 6. Conclusion

This EDA has provided a comprehensive understanding of the `Crime_Data.csv` dataset, highlighting prevalent crime types, the distribution of associated 'Commits', and their temporal dynamics. The findings can serve as a foundation for more in-depth investigations, predictive modeling, or informing crime prevention strategies by identifying critical areas and periods.

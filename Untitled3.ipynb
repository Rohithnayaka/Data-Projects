## 5. Summary Report of Technical Indicators

We have calculated and visualized several key technical indicators to gain a deeper understanding of Ferrari N.V. (RACE) stock's behavior:

1.  **Daily Returns (`Daily_Return`)**:
    *   **Purpose**: Measures the percentage change in the stock's closing price from one day to the next.
    *   **Interpretation**: Quantifies daily price movements and serves as a fundamental measure of performance and risk. Positive values indicate gains, negative values indicate losses.

2.  **20-Day Rolling Average of Daily Returns (`Daily_Return_Rolling_Avg_20d`)**:
    *   **Purpose**: Smoothes out short-term fluctuations in daily returns to reveal underlying trends.
    *   **Interpretation**: A positive rolling average suggests sustained upward momentum in returns, while a negative one suggests sustained downward momentum. It provides a less volatile view compared to raw daily returns.

3.  **20-Day Rolling Volatility (`Daily_Volatility_20d`)**:
    *   **Purpose**: Measures the standard deviation of daily returns over a 20-day period.
    *   **Interpretation**: Indicates the stock's risk level or price fluctuation. Higher volatility suggests greater price swings and higher risk; lower volatility suggests more stable price movements.

4.  **Relative Strength Index (RSI)**:
    *   **Purpose**: A momentum oscillator that measures the speed and change of price movements.
    *   **Interpretation**: Ranges from 0 to 100. Typically, an RSI above 70 suggests the stock is **overbought** (potentially due for a pullback), and an RSI below 30 suggests it is **oversold** (potentially due for a rebound).

5.  **Moving Average Convergence Divergence (MACD)**:
    *   **Purpose**: A trend-following momentum indicator that shows the relationship between two exponential moving averages (EMAs) of a stock's price.
    *   **Interpretation**: Consists of:
        *   **MACD Line**: Difference between 12-period EMA and 26-period EMA. Indicates momentum.
        *   **Signal Line**: 9-period EMA of the MACD Line. Acts as a trigger for buy/sell signals.
        *   **MACD Histogram**: Difference between MACD Line and Signal Line. Shows strength of momentum.
        *   Crossovers of the MACD and Signal lines, and movements relative to the zero line, are used to identify potential trend changes and buy/sell signals.

6.  **Bollinger Bands (`SMA_20`, `Upper_Band`, `Lower_Band`)**:
    *   **Purpose**: Measures market volatility and provides potential dynamic support and resistance levels.
    *   **Interpretation**: Consists of a 20-day Simple Moving Average (SMA) as the mid-band, and an upper and lower band typically two standard deviations away from the SMA.
        *   **Band Width**: Expands during high volatility, contracts during low volatility.
        *   **Price Action**: Prices tend to revert to the SMA. Touching the upper band can suggest **overbought** conditions, while touching the lower band can suggest **oversold** conditions.
        *   **Breakouts**: A price moving outside the bands can signal a significant price move.

```python
# Calculate 20-day Simple Moving Average (SMA)
df['SMA_20'] = df['Close'].rolling(window=20).mean()

# Calculate 20-day Standard Deviation
df['STD_20'] = df['Close'].rolling(window=20).std()

# Calculate Upper Bollinger Band (Upper Band)
df['Upper_Band'] = df['SMA_20'] + (df['STD_20'] * 2)

# Calculate Lower Bollinger Band (Lower Band)
df['Lower_Band'] = df['SMA_20'] - (df['STD_20'] * 2)

# Display the last few rows with the new Bollinger Band columns
display(df.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Close Price', color='blue', alpha=0.7)
plt.plot(df.index, df['SMA_20'], label='20-Day SMA', color='orange')
plt.plot(df.index, df['Upper_Band'], label='Upper Bollinger Band', color='red', linestyle='--')
plt.plot(df.index, df['Lower_Band'], label='Lower Bollinger Band', color='green', linestyle='--')
plt.fill_between(df.index, df['Lower_Band'], df['Upper_Band'], color='gray', alpha=0.1)
plt.title('Bollinger Bands for Ferrari N.V. (RACE)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate the 12-period EMA
df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()

# Calculate the 26-period EMA
df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()

# Calculate MACD Line
df['MACD'] = df['EMA_12'] - df['EMA_26']

# Calculate Signal Line (9-period EMA of MACD)
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()

# Calculate MACD Histogram
df['MACD_Histogram'] = df['MACD'] - df['Signal_Line']

# Display the last few rows with the new MACD columns
display(df.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
plt.plot(df.index, df['MACD'], label='MACD Line', color='blue')
plt.plot(df.index, df['Signal_Line'], label='Signal Line', color='red')
plt.bar(df.index, df['MACD_Histogram'], label='MACD Histogram', color='gray', alpha=0.7)
plt.title('MACD Indicator Over Time')
plt.xlabel('Date')
plt.ylabel('MACD Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate daily price changes
delta = df['Close'].diff()

# Get positive and negative changes
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

# Calculate exponential moving average for gain and loss
window = 14 # Standard RSI window
avg_gain = gain.ewm(com=window - 1, adjust=False).mean()
avg_loss = loss.ewm(com=window - 1, adjust=False).mean()

# Calculate Relative Strength (RS)
rs = avg_gain / avg_loss

# Calculate RSI
df['RSI'] = 100 - (100 / (1 + rs))

# Display the first few rows with the new RSI column
display(df.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
sns.lineplot(x=df.index, y=df['RSI'], color='green')
plt.title('Relative Strength Index (RSI) Over Time')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
plt.axhline(30, color='blue', linestyle='--', label='Oversold (30)')
plt.grid(True)
plt.legend()
plt.show()
```
